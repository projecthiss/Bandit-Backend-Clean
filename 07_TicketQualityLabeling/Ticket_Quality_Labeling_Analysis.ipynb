```python
import pandas as pd
from sklearn.metrics import cohen_kappa_score
```

```python
df = pd.read_excel("/content/gdrive/My Drive/Colab Notebooks/Pipeline_Extended/output_ticket_labeling.xlsx", sheet_name=0, na_values="NaN")
df['problem_label_annotator 1'].astype('int', errors='ignore')
df['problem_label_annotator 2'].astype('int', errors='ignore') 
df['solution_label_annotator 1'].astype('int', errors='ignore') 
df['solution_label_annotator 2'].astype('int', errors='ignore') 
```

```python
col_one_list = df['problem_label_melina'].tolist()
col_two_list = df['problem_label_philipp'].tolist()
cohen_kappa = cohen_kappa_score(col_one_list, col_two_list)
print("-------Problem Description-------")
print("Common sense: ", df.shape[0]-df['problem_labels_merged'].value_counts()[-1])
print("Cohens kappa for Problem: ", cohen_kappa)
df['problem_labels_merged'].value_counts()
```

```python
col_one_list = df['solution_label_melina'].tolist()
col_two_list = df['solution_label_philipp'].tolist()
cohen_kappa = cohen_kappa_score(col_one_list, col_two_list)
print("-------Solution Description-------")
print("Common sense: ", df.shape[0]-df['solution_labels_merged'].value_counts()[-1])
print("Cohens kappa for Solution: ", cohen_kappa)
df['solution_labels_merged'].value_counts()
```
